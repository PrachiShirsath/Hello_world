```python
import pandas as pd
from google.colab import files

# Upload dataset
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Load dataset
df = pd.read_csv(filename)

print("Dataset Loaded Successfully")
print(df.head())
```

```python
#data cleaning:
# Check missing values
print("Missing values:\n", df.isnull().sum())

# Drop missing values
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

print("Cleaned Dataset Shape:", df.shape)
```

```python
#Simple linear regression:
from sklearn.linear_model import LinearRegression

X = df[['danceability']]
y = df['popularity']

slr = LinearRegression()
slr.fit(X, y)

print("Prediction (danceability = 0.6):",
      slr.predict([[0.6]]))
```

```python
# Multiple linear regression:
X_multi = df[['danceability', 'energy', 'tempo']]
y_multi = df['popularity']

mlr = LinearRegression()
mlr.fit(X_multi, y_multi)

print("Multiple LR Prediction:",
      mlr.predict([[0.6, 0.7, 120]]))
```

```python
#Logistic regression:
# Create classification target
df['popular'] = (df['popularity'] >= df['popularity'].median()).astype(int)

from sklearn.linear_model import LogisticRegression

X_log = df[['danceability', 'energy']]
y_log = df['popular']

log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_log, y_log)

print("Logistic Probability:",
      log_reg.predict_proba([[0.6, 0.7]]))
print("Logistic Prediction:",
      log_reg.predict([[0.6, 0.7]]))
```

```python
#Decision tree:
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_log, y_log)

print("Decision Tree Prediction:",
      dt.predict([[0.65, 0.8]]))
```

```python
#Support vector machine:
from sklearn.svm import SVC

svm = SVC(probability=True)
svm.fit(X_log, y_log)

print("SVM Prediction:",
      svm.predict([[0.65, 0.8]]))
```

```python
#Random forest Implementation :
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_log, y_log)

print("Random Forest Prediction:",
      rf.predict([[0.6, 0.75]]))
```

```python
#Gradient Descent:
import numpy as np

X_gd = df['danceability'].values
y_gd = df['popularity'].values

m = 0
c = 0
lr = 0.01
epochs = 1000
n = len(X_gd)

for _ in range(epochs):
    y_pred = m * X_gd + c
    dm = (-2/n) * sum(X_gd * (y_gd - y_pred))
    dc = (-2/n) * sum(y_gd - y_pred)
    m -= lr * dm
    c -= lr * dc

print("Gradient Descent Equation:")
print("Popularity = {:.4f} * Danceability + {:.4f}".format(m, c))
print("Prediction:", m*0.6 + c)
```

```python
#⁠Neural Networks:
from tensorflow import keras
from tensorflow.keras import layers

X_nn = df[['danceability', 'energy', 'tempo']]
y_nn = df['popularity']

model = keras.Sequential([
    layers.Dense(16, activation='relu', input_shape=(3,)),
    layers.Dense(8, activation='relu'),
    layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')

model.fit(X_nn, y_nn, epochs=50, verbose=0)

import numpy as np
test_input = np.array([[0.6, 0.7, 120]], dtype=float)

print("Neural Network Prediction:",
      model.predict(test_input))
```
